```python
import matplotlib.pyplot as plt
plt.plot([1,2,4,9,5,3])
plt.show()
```

It is as simple as calling  the plot function with some data and then calling the show function!!

Of the plot function is given one array of data, It will use it as  the coordinate on the vertical axis and it will just use  point's Index in the array as the horizontal coordinate . I can also provide two arrays : onw for the horizontal axis x, and the second for the vertical axis y:


```python
plt.plot([-3,-2,5,0],[1,6,4,3])
plt.show()
```

```python
plt.plot([-3,-2,5,0],[1,6,4,3])
plt.axis([-4,6,0,7])
plt.show()
```

```python
import numpy as np
x = np.linspace(-2,2,500)
y = x**2

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("square function")
plt.xlabel("x")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```

```python
x = np.linspace(-1.4,1.4,30)
plt.plot(x,x,'g--',x,x**2,'r:',x,x**3,'b^')
plt.show()
```

```python

x = np.linspace(-1.4, 1.4, 30)
line1, line2, line3 = plt.plot(x, x, 'g--', x, x**2, 'r:', x, x**3, 'b^')
line1.set_linewidth(3.0)
line1.set_dash_capstyle("round")
line3.set_alpha(0.2)
plt.show()
```

```python
#saving  a figure
x = np.linspace(-1.4,1.4,30)
plt.plot(x,x**2)
plt.savefig("my_square_function.png",transparent = True)
```

```python
%matplotlib nbagg
import matplotlib.animation as animation
```

```python
x = np.linspace(-1, 1, 100)
y = np.sin(x**2*25)
data = np.array([x, y])

fig = plt.figure()
line, = plt.plot([], [], "r-") # start with an empty plot
plt.axis([-1.1, 1.1, -1.1, 1.1])
plt.plot([-0.5, 0.5], [0, 0], "b-", [0, 0], [-0.5, 0.5], "b-", 0, 0, "ro")
plt.grid(True)
plt.title("Marvelous animation")

# this function will be called at every iteration
def update_line(num, data, line):
    line.set_data(data[..., :num] + np.random.rand(2, num) / 25)  # we only plot the first `num` data points.
    return line,

line_ani = animation.FuncAnimation(fig, update_line, frames=100, fargs=(data, line), interval=67)
plt.show()
```

```python
Writer = animation.writers['ffmpeg']
writer = Writer(fps=15, metadata=dict(artist='Me'), bitrate=1800)
line_ani.save('my_wiggly_animation.mp4', writer=writer)
```
